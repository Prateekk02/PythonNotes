### merge()
- Does work on values.

```python
data1 = pd.DataFrame({ 'key1' : [1,2,4,6,6],
                        'key2' : [5,5,6,7,8],
                        'key3' : [3,5,5,6,6]})
```

```python
data1
```

```python
data2 = pd.DataFrame({ 'key1' : [1,2,45,6,67],
                        'key2' : [56,5,6,7,8],
                        'key3' : [3,56,5,6,6]})
```

```python
data2
```

```python
pd.merge(data1, data2)  # does inner merge by default.
```

```python
pd.merge(data1, data2, how = 'left') # Doing left merge
```

```python
pd.merge(data1,data2, how = 'right')
```

```python
pd.merge(data1, data2, how = 'outer')
```

```python
pd.merge(data1, data2, how = 'cross')
```

### apply()

- Analogous to map in filter,reduce,map section.

```python
df
```

We want to convert Fare from dollar to rupees

```python
df['Fare_INR'] = df['Fare'].apply(lambda x : x*80)
```

```python
df.head()
```

```python
def euro_to_inr(x):
    return (x*80)

df['Fare_INR'] = df['Fare'].apply(euro_to_inr)
```

```python
df
```

Now we want to store length of names in a saparate column

```python
df['Len_of_name'] = df['Name'].apply(len)
```

```python
df.head()
```

Now we want to write cheapest,mid and high corresponding to that column of fare around 10 

```python
def cat_fare(x):
    if x<10:
        return 'cheap'
    elif x>=10 and x<20:
        return 'mid'
    else:
        return 'high'   
```

```python
df['cat_fare'] = df['Fare'].apply(cat_fare)
```
